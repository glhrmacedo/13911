### Questão 11

```python
dataset['critical_ratio'] = dataset['icu_patients'] / dataset['hosp_patients']
dataset = dataset.drop(columns = ['hosp_patients'])
dataset.sort_values('critical_ratio', ascending = False)['date'].loc[0]
```

### Questão 12

```python
dataset.groupby('location').filter(lambda row : row['new_cases'].mean() > 10000)['location'].unique()
```

### Questão 13

```python
dataset.groupby('location').filter(lambda row : row['icu_patients'].std() > 1000)['location'].unique()
```

### Questão 14

```python
dataset['death_population_ratio'] = dataset['total_deaths'] / dataset['population']
dataset.groupby('location').filter(lambda row : row['death_population_ratio'].max() > 0.02)['diabetes_prevalence'].max()
```
